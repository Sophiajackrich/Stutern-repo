**b.** Write a NumPy code that calculates: 

$$
result = \frac{\sum{((x - \bar{x}) * (y - \bar{y}))}}{\sum{(x - \bar{x})^2}}

`X` and `y` are given below:

```python
rng = np.random.default_rng(seed=5)
```

````python
# Write your solution here
# Qs 1a: Showing the properties of question 1a
x = np.linspace(0, 10, 10001)
show_array_properties(arr=x)

# Solution
step_1 =  np.square(-x/10)
step_2 = np.sin(x)
# final step
y = np.exp(np.square(-x/10)) * np.sin(x)
print("The solution for question 1a is:", y)
````

````python
# Write your solution here
# Qs 1b
# reading out the data
rng = np.random.default_rng(seed=5)
X = rng.random(size=(35, 2), dtype=np.float32)
y = rng.random(size=(35, 1), dtype=np.float32)
# checking properties of the X and y array
show_array_properties(arr=X)
show_array_properties(arr=y)
````


$$
result = \frac{\sum{((x - \bar{x}) * (y - \bar{y}))}}{\sum{(x - \bar{x})^2}}
$$

````python
# Getting the X and y mean of Question 1b
X_mean = np.mean(X)
print(X_mean)

y_mean = np.mean(y)
print(y_mean)
````

````python
# Question 1b Solution
numerator = np.sum((X - X_mean) * (y - y_mean))
denominator = np.sum(np.square(X - X_mean))

result = numerator / denominator
result
````
