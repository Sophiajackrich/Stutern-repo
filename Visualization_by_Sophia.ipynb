**QUESTION 1**: 

Given the following dataset:

a = np.random.normal(loc=64, scale=2, size=500)

b = np.random.normal(loc=70, scale=2, size=100)

Write the appropriate to display multiple histogram on the same axis. The plots for both datasets show be fully
visible.
[Hint: Use the 'alpha' argument]


```python
# Code
a = np.random.normal(loc=64, scale=2, size=500)

b = np.random.normal(loc=70, scale=2, size=100)
print(a[:3])
print(b[:3])

# Plot multiple hist on same axis
#include a figsize
plt.figure(figsize=[6,4])
plt.hist(a, bins=15, label="A", color="b", alpha=0.5, rwidth=0.8, density=True)
plt.hist(b, bins=15, label="B", color="orange", alpha=0.5, rwidth=0.8, density=True)
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.title("Multiple Histogram of A & B")
plt.legend();
```

**QUESTION 3:**

How would you use matplotlib module to make two different subplots on different axes with the following 
characteristics?

a) First subplot: 

i. pie chart using the 'sex' column. Ensure the percentage on the pie chart is taken to 2 d.p.

b) Second subplot:

i. Create a bar-chart with the 'region' column. 

ii. Carefully label the chart with x-axis: 'Region'; y-axis: 'Counts'; title: 'Bar plot for Region'.

```python
# First Subplot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

df['sex'].value_counts().plot(kind='pie', autopct="%.2f", textprops={'fontsize': 20}, 
                              colors=['violet', 'lime'], ax=ax1)
print(type(ax1))


# Second plot
df['region'].value_counts().plot(kind='bar', ax=ax2)
ax2.set_xlabel("Region", size=22)
ax2.set_ylabel('Counts', size=22)
ax2.set_title('Bar plot for Region', size=22)
plt.tight_layout()
plt.savefig('Barplot.png', bbox_inches='tight', dpi=150);

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
print(type(ax1))
```

**QUESTION 5:**

Using the dataset given below,

x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)

y_var = np.random.normal(loc = 50, scale = 1.5, size = 300) 

z_var = np.random.normal(loc = 60, scale = 2, size = 300)

i. Create histograms for the two variables,x_var and y_var, on the same axis. Ensure the two plots can easily 
be see using the appropriate parameter. Also, use the legend() function for the plot.

ii. Make two subplots using the 3 variables as follows:

1. Make a lineplot of y_var against z_var for the first subplot
2. Make a histogram for the second subplot using x_var.
[Hint: All plots should be well labeled] 

```python
x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300)
z_var = np.random.normal(loc = 60, scale = 2, size = 300)

#i. Plot Multiple Histograms
plt.hist(x_var, bins=10, label="X var", rwidth=0.8,  alpha=0.8, density=True)
plt.hist(y_var, bins=10, label="Y var", rwidth=0.8, alpha=0.8, density=True)
plt.xlabel("Variable: X")
plt.ylabel("Variable: Y")
plt.title("Multiple Histogram: X_VAR & Y_VAR ON SAME AXES")
plt.legend();
```

```python
# ii. Make two subplots using the 3 variables as follows:
# 1. Make a lineplot of y_var against z_var for the first subplot
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=[10,4])
ax1.plot(y_var, z_var, color='C1', alpha=0.4)
ax1.set_xlabel("Y Variable")
ax1.set_ylabel("Z Variable")
ax1.set_title("LinePlot of y_var against z_var")

# 2. Make a histogram for the second subplot using x_var. 
plt.hist(x_var, bins=15, color="#AAAAAA", rwidth=0.8)
ax2.set_title("Single Histogram Plot of X Variable");

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
print(type(ax1))
print(type(ax2))
```
