```python
# there are less missing values in Gender,Married,Dependents,Credit_History, and Self_Employed features so we can fill them using the mode of the features.

df3['Gender'].fillna(df3['Gender'].mode()[0], inplace=True)
df3['Married'].fillna(df3['Married'].mode()[0], inplace=True)
df3['Dependents'].fillna(df3['Dependents'].mode()[0], inplace=True)
df3['Self_Employed'].fillna(df3['Self_Employed'].mode()[0], inplace=True)
df3['Credit_History'].fillna(df3['Credit_History'].mode()[0], inplace=True)
```

```python
# we will look at the value count of the Loan amount term variable.
df3['Loan_Amount_Term'].value_counts()
```

It can be seen that in the loan amount term variable, the value of 360 is repeated the most, So we will replace the missing values in this variable in the variable using the mode of this variable.

```python
df3['Loan_Amount_Term'].fillna(df3['Loan_Amount_Term'].mode()[0], inplace=True)
```

```python
# using the median to fill the null values
df3['LoanAmount'].fillna(df3['LoanAmount'].median(), inplace=True)
```

```python
# let's check whether all the missing values are filled in the dataset
df3.isnull().sum()
```

# Model Building

```python
# let's drop the Loan_ID variable as it does not have any effect on the loan status, same with the test dataset
df3=df3.drop('Loan_ID',axis=1)
df5=df5.drop('Loan_ID',axis=1)
```

```python
# Sklearn require the target variable in a separate daaset. so,we will drop our target variable from the training dataset and save it in another dataset.
X = df5.drop('Loan_Status',1)
y = df5.Loan_Status
```

Now we will make dummy variables for the categorical variables. dummy variable turns categorical variables into a series of 0 and 1, making them a lot easier to quantify and compare.

```python
X=pd.get_dummies(X)
df5=pd.get_dummies(df5)
df3=pd.get_dummies(df3)
```

Now we will train the model on the training dataset and make predictions for the test dataset. But can we validate these predictions? one way of doing this is to divide our train dataset into two parts: train and validation. we can train the model on this training part and using that make predictions for the validation part. in this way, we can validate our predictions as we have true predictions for the validation part (which we do not have for the test dataset).

we will use the train_test_split function from sklearn to divide our train dataset. So first, let us import train_test_split.

```python
df5.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_cv, y_train, y_cv = train_test_split(X,y, test_size =0.3)
```

The dataset has been divided into trainingand validation parts. let us import LogisticRegression and accuracy_score from sklearn and fit the logistic regression model.

```python

from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
# Generate a sample dataset with 7 features
X, y = make_classification(n_samples=100, n_features=7, random_state=42)
```

```python
# Adapt the Random Forest classifier
if X.shape[1] == 7 and X.shape[1] < 50:
    # Reduce the number of estimators (trees) to account for the smaller feature set
    rf = RandomForestClassifier(n_estimators=10, random_state=42)
else:
    # Use a different classifier, such as Logistic Regression
    rf = LogisticRegression(random_state=42)
```

In this code snippet, if the dataset has 7 features (as checked by X.shape[1] == 7) and is less than 50, the code adapts the Random Forest classifier by reducing the number of estimators (n_estimators=10). Otherwise, it falls back to using a different classifier, in this case, Logistic Regression.

```python
# Fit the classifier to the data
rf.fit(X, y)
```

```python
pred=rf.predict(X)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(pred,y))
print('\n')
print(confusion_matrix(pred,y))
```
