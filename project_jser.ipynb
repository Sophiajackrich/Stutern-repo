````python
# Using one hot encoder on the categorical columns
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

# le = OneHotEncoder()
le = LabelEncoder()
cat_cols = ['Workclass', 'Education', 'Marital_status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Country']
#OH_X = pd.DataFrame(le.fit_transform(salary[cat_cols]).toarray())
for i in cat_cols:
    salary[i] = le.fit_transform(salary[i])
    
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
salary[["Age", "Final_weight", "Education_num", "Capital_gain", "Capital_loss", "Hours_per_week"]] = scaler.fit_transform(salary[["Age", "Final_weight", "Education_num", "Capital_gain", "Capital_loss", "Hours_per_week"]])


X = salary.drop("Salary", axis=1)
#X = X.drop(cat_cols, axis = 1)
#X = X.join(OH_X)
y = salary.Salary.apply(lambda x: 1 if x == ">50K" else 0)
#y = salary["Salary"]
#X.columns = X.columns.astype(str)
X.shape, y.shape

# one_hot_encoder = OneHotEncoder(handle_unknown = "ignore", sparse_output = Fals
# one_hot_encoder = OneHotEncoder()
# OH_X = pd.DataFrame(one_hot_encoder.fit_transform(X[cat_cols]))
# OH_X.index = X.index
# # Normalizing the numerical columm
# normalizer = StandardScaler()
# norm_X = pd.DataFrame(normalizer.fit_transform(X[num_cols]))
# norm_X.columns = X[num_cols].columns
# norm_X.index = X.index
# prep_X = pd.concat([norm_X, OH_X], axis = 1)
# prep_X.columns = prep_X.columns.astype(str)
````

````python
# Splitting the dataset into Train and test dataset
X_train, X_test , y_train, y_test = train_test_split(X, y, train_size = 0.7)
````

````python
# having a view of the test set
X_test.head()
````

````python
# modelling
random_forest = RandomForestClassifier(n_estimators = 90)
random_forest.fit(X_train, y_train)
Y_pred = random_forest.predict(X_test)
mse = mean_squared_error(y_test, Y_pred, squared = False)
f1 = f1_score(y_test, Y_pred, average = 'weighted')
recall = recall_score(y_test, Y_pred, average = 'weighted')
print("The following below are the evaluation metrics for the model ")
print("The model had a mean squared error of {}".format(mse))
print("The model had an f1 score of {}".format(f1))
print("The model had a recall of {}".format(recall))
````
